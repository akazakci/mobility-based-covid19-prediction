```python
import pandas as pd
```

# Intra-departmental activity

```python
intra =  pd.read_csv('intra_dept_by_date.csv', index_col = 0)
```

```python
intra.sort_values('date', inplace=True)
intra
```

# Inter-departmental activity

```python
inter = pd.read_csv('influx_to_dept_by_date.csv', index_col = 0)
```

```python
inter.head()
```

```python
inter.code_insee.unique()
```

# Covid incidence data

```python
incidences = pd.read_csv('incidence.csv', index_col = 0)
```

```python
incidences.head(3)
```

```python
incidences.shape
```

```python
incidences.code_insee.unique().shape
```

The following departments are not represented in the features we built from original Facebook data.

```python
set(incidences.code_insee.unique()) - set(inter.code_insee.unique())
```

# We only have data for 95 departments, so we eliminate missing departments from incidences.

```python
incidences = incidences[incidences['code_insee'].isin(inter.code_insee.unique())]
```

```python
incidences.code_insee.unique().shape
```

# Merge Intra Features

```python
incidences = pd.merge(incidences, intra, how='outer', on = ['code_insee','date'])
```

```python
incidences.shape
```

```python
incidences.head(3)
```

# Merge inter to new incidences (incidences + intra)

```python
incidences = pd.merge(incidences, inter, how='outer', on = ['code_insee','date'])
```

```python
incidences.shape
```

```python
incidences.head(3)
```

```python
incidences.columns
```

# Merge (Dynamic) Graph Statistics to last version of incidences

```python
graph_stats = pd.read_csv('graph_stats_dept_by_date.csv', index_col = 0)
graph_stats.head()
```

```python
incidences = pd.merge(incidences, graph_stats, how='outer', on = ['code_insee','date'])
```

```python
incidences.shape
```

```python
incidences.columns
```

# Imputation strategy

```python
# I will not use the code below, since I think filling with 0s is a better option for the above reason
#from sklearn.impute import SimpleImputer
#imputer = SimpleImputer()
#data_with_imputed_values = imputer.fit_transform(some_data)
```

Thus:

```python
incidences = incidences.fillna(0)
```
